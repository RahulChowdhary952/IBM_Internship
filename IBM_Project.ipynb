```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in the CSV file
df = pd.read_csv('SampleSuperstore.csv')
df
```

```python
df.head()
```

 # **Preprocessing**

```python
df.describe()
```

```python
df.info()
```

#### Handle missing values

```python

df = df.fillna(0)

df
```

```python
# Normalize monetary columns
df['Sales'] = df['Sales'] / df['Sales'].max()
df['Profit'] = df['Profit'] / df['Profit'].max()
df
```

# **Analytics**

```python
# Group by Region and calculate average profit
print(df.groupby('Region')['Profit'].mean())
```

```python

average_profit_by_region = df.groupby('Region')['Profit'].mean()


plt.figure(figsize=(6, 6))
average_profit_by_region.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.title('Average Profit by Region')
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Sales', 'Profit']].corr()

#heatmap
plt.figure(figsize=(6, 2))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap: Sales vs Profit')
plt.show()
```

```python
# Scatter plot of Sales vs Profit
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()

```

```python
# Histogram of Region
df['Region'].value_counts().plot.bar()
plt.title('Orders by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()

```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
average_profit_by_region.plot(kind='bar', color='skyblue', ax=ax1)

ax1.set_xlabel('Region')
ax1.set_ylabel('Average Profit')
ax1.set_title('Average Profit by Region')


df['Region'].value_counts().plot(kind='bar', ax=ax2)

ax2.set_xlabel('Region')
ax2.set_ylabel('Count')
ax2.set_title('Orders by Region')

plt.tight_layout()
plt.show()
```

```python
# Boxplot of Profit by Category
df.boxplot(column='Profit', by='Category')
plt.ylabel('Profit')
plt.title('Profit by Category')
plt.show()
```

```python

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Sales by Category
sns.barplot(x='Category', y='Sales', data=df, ax=axs[0,0])
axs[0,0].set_title('Sales by Category')

# Profit by Segment
sns.boxplot(x='Segment', y='Profit', data=df, ax=axs[0,1])
axs[0,1].set_title('Profit by Segment')

# Sales by Region
sns.countplot(x='Region', data=df, ax=axs[1,0])
axs[1,0].set_title('Number of Orders by Region')

# Scatterplot of Sales vs Profit
sns.scatterplot(x='Sales', y='Profit', data=df, ax=axs[1,1])
axs[1,1].set_title('Sales vs Profit')

plt.tight_layout()
plt.show()
```
